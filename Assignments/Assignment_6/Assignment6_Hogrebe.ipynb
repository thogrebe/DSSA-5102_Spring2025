```python
# Create our first table in the database file using SQL statements:
# We want our table column names to match what is in the .csv file
from sqlalchemy import text

first_table_query = """
CREATE TABLE IF NOT EXISTS students (
    Student_ID VARCHAR(10),
    Age INT,
    Gender VARCHAR(10),
    High_School_GPA FLOAT,
    SAT_Score INT,
    University_Ranking INT,
    University_GPA FLOAT,
    Field_of_Study VARCHAR(50),
    Internships_Completed INT,
    Projects_Completed INT,
    Certifications INT,
    Soft_Skills_Score INT,
    Networking_Score INT,
    Job_Offers INT,
    Starting_Salary FLOAT,
    Career_Satisfaction INT,
    Years_to_Promotion INT,
    Current_Job_Level VARCHAR(20),
    Work_Life_Balance INT,
    Entrepreneurship VARCHAR(5)
);
"""
# Note that the primary key for this table is a column/field "id"
# This is not a field that existed previously. AUTO_INCREMENT automatically generates a unique value for each new row added to the table. 
# Each new value is one greater than the previous value. We cannot make the Date column/field our primary key, because it is not unique.
```

```python
#Execute the query:
# Execute the query
with engine.connect() as connection:
    connection.execute(text(first_table_query))

print("✅ First table created successfully!")
```

Define your SQL data types for your first table: <br><br>
**My SQL data types for my first table:**

- `Student_ID`: VARCHAR(10) – It's a unique identifier but not numeric, so we use a short string.
- `Age`: INT – Age is a whole number.
- `Gender`: VARCHAR(10) – String field with values like Male, Female, or Other.
- `High_School_GPA`: FLOAT – GPA has decimal values.
- `SAT_Score`, `University_Ranking`, etc.: INT – Standard integer values.
- `Field_of_Study`, `Current_Job_Level`, `Entrepreneurship`: VARCHAR – These are descriptive text fields.
- `Starting_Salary`, `University_GPA`: FLOAT – Salary and GPA often contain decimals.

**Primary Key:**
I did not assign a formal primary key in this assignment because `Student_ID` is already a unique identifier. If needed, it could be designated as the primary key.

```python
engine = create_engine("sqlite:///education_local.db")
```

```python
from sqlalchemy import text

with engine.connect() as connection:
    connection.execute(text("""
        CREATE TABLE IF NOT EXISTS students (
            Student_ID TEXT,
            Age INTEGER,
            Gender TEXT,
            High_School_GPA FLOAT,
            SAT_Score INTEGER,
            University_Ranking INTEGER,
            University_GPA FLOAT,
            Field_of_Study TEXT,
            Internships_Completed INTEGER,
            Projects_Completed INTEGER,
            Certifications INTEGER,
            Soft_Skills_Score INTEGER,
            Networking_Score INTEGER,
            Job_Offers INTEGER,
            Starting_Salary FLOAT,
            Career_Satisfaction INTEGER,
            Years_to_Promotion INTEGER,
            Current_Job_Level TEXT,
            Work_Life_Balance INTEGER,
            Entrepreneurship TEXT
        )
    """))
print("✅ SQLite table created successfully!")
```

#### Creating a local database (.db file) from scratch in Python using SQLite:<br>


```python
# Load necessary packages:
from sqlalchemy import create_engine, inspect, text # Database navigation
import sqlite3 # A second option for working with databases
import pandas as pd # Python data manilpulation
```

```python
# Load necessary packages
from sqlalchemy import create_engine
import pandas as pd
import sqlite3

# Define the name of the SQLite .db file
db_file = "education_local.db"
engine = create_engine(f"sqlite:///{db_file}")

# Load the CSV again (if not already loaded)
df = pd.read_csv("education_career_success.csv")

# Write the DataFrame to the SQLite database
df.to_sql(name="students", con=engine, if_exists="replace", index=False)

print("✅ Data written to local SQLite .db file successfully!")
```

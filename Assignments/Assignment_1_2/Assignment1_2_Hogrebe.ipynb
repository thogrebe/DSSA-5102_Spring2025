```python
import re  # Regular expressions for column name cleaning

# Convert all column names to lowercase to maintain consistency
df_drugs.columns = df_drugs.columns.str.lower()

# Replace spaces with underscores to make column names SQL-friendly
df_drugs.columns = df_drugs.columns.str.replace(r'\s+', '_', regex=True)

# Remove special characters to prevent SQL errors
df_drugs.columns = df_drugs.columns.str.replace(r'[^a-z0-9_]', '', regex=True)

# Print updated column names to verify changes
print("\nUpdated column names:")
print(df_drugs.columns.tolist())

# Display missing values summary
print("\nMissing Values Summary:")
print(df_drugs.isna().sum())

# Drop columns where more than 50% of values are missing
threshold = len(df_drugs) * 0.5  # Set threshold (50% of dataset size)
df_drugs = df_drugs.dropna(axis=1, thresh=threshold)

# Print remaining columns after dropping high-missing-value columns
print("\nColumns remaining after dropping high-missing-value columns:")
print(df_drugs.columns.tolist())

# Fill missing numerical values with the median of the column
num_cols = df_drugs.select_dtypes(include=[np.number]).columns  # Identify numerical columns
df_drugs[num_cols] = df_drugs[num_cols].apply(lambda x: x.fillna(x.median()))

# Fill missing categorical values with the mode (most frequently occurring value)
cat_cols = df_drugs.select_dtypes(include=['object']).columns  # Identify categorical columns
df_drugs[cat_cols] = df_drugs[cat_cols].apply(lambda x: x.fillna(x.mode()[0]))

# Display unique values in 'age' column to check if we need to remove any categories
print("\nUnique Age Groups Before Cleaning:", df_drugs['age'].unique())

# Drop 'All ages' as it aggregates all groups and may skew analysis
df_drugs = df_drugs[df_drugs['age'] != 'All ages']

# Verify the 'age' column after removal
print("\nUnique Age Groups After Cleaning:", df_drugs['age'].unique())

# Convert 'year' column to integer if it is stored as a float or string
df_drugs['year'] = pd.to_numeric(df_drugs['year'], errors='coerce').astype('Int64')

# Convert numerical columns to appropriate numeric types if they are stored as strings
for col in num_cols:
    df_drugs[col] = pd.to_numeric(df_drugs[col], errors='coerce')

# Check for duplicate rows
duplicate_count = df_drugs.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicate_count}")

# Drop duplicate rows if any exist
df_drugs = df_drugs.drop_duplicates()

# Display basic statistics to identify potential outliers
print("\nSummary statistics for numerical columns:")
print(df_drugs.describe())

# Cap extreme values in 'estimate' column at 99th percentile to avoid skewing analysis
upper_limit = df_drugs['estimate'].quantile(0.99)  # Get 99th percentile value
df_drugs['estimate'] = np.where(df_drugs['estimate'] > upper_limit, upper_limit, df_drugs['estimate'])

# Define the output file name for the cleaned dataset
cleaned_file_path = "drug_overdose_cleaned.csv"

# Save the cleaned dataset as a new CSV file
df_drugs.to_csv(cleaned_file_path, index=False)

# Display confirmation message
print("\nData cleaning completed! Cleaned dataset saved as:", cleaned_file_path)
```

# **Dataset Documentation: Drug Overdose Death Rates by Drug Type, Sex, Age, Race, and Hispanic Origin – United States**

## **1. Where is the data from?**
The dataset is provided by the **U.S. Department of Health & Human Services (HHS)** and is available on [Data.gov](https://catalog.data.gov/dataset/drug-overdose-death-rates-by-drug-type-sex-age-race-and-hispanic-origin-3f72f).

## **2. How was it collected?**
The data was collected from **national health records**, including death certificates and medical examiner reports compiled by the **Centers for Disease Control and Prevention (CDC)** and the **National Center for Health Statistics (NCHS)**.

## **3. How was it extracted?**
The dataset was extracted from CDC’s mortality surveillance systems, using standard reporting protocols to aggregate state-level and national-level overdose death rates. It was then made available on Data.gov in machine-readable formats such as CSV and JSON.

## **4. What program was used to clean the data?**
The dataset was cleaned using **Python (Pandas, NumPy, and Regular Expressions)** in a Jupyter Notebook 

## **5. How was the data cleaned or transformed?**
- **Standardized Column Names:**
  - Converted all column names to lowercase.
  - Replaced spaces with underscores to make them SQL-friendly.
  - Removed special characters to prevent SQL errors.

- **Handling Missing Data:**
  - Identified missing values across all columns.
  - Dropped columns where more than **50% of values were missing**.
  - Filled missing numerical values with the **median** of the column.
  - Filled missing categorical values with the **mode** (most frequently occurring value).

- **Filtering & Transforming Data:**
  - **Removed the "All ages" category** from the `age` column to avoid aggregation bias.
  - Converted the `year` column to an integer type.
  - Ensured numerical columns were stored in numeric format instead of strings.

- **Removing Duplicates & Handling Outliers:**
  - Identified and removed duplicate rows.
  - Capped extreme values in the `estimate` column at the **99th percentile** to prevent skewing analysis.

- **Final Steps:**
  - Saved the cleaned dataset as `drug_overdose_cleaned.csv` for further analysis.

## **6. What are the units of the numeric data?**
- **Death rates are measured in deaths per 100,000 resident population**.
- **Numerical identifiers** (e.g., `panel_num`, `unit_num`) represent categorical classifications.

## **7. What were the formulas used in column creation(s)?**
- **Crude rate formula:**
  \[
  \text{Crude Rate} = \left(\frac{\text{Number of deaths in a group}}{\text{Total population of that group}}\right) \times 100,000
  \]
- **Age-adjusted rate formula:**
  \[
  \sum_{i} \left(\frac{\text{Age group deaths}_i}{\text{Age group population}_i}\right) \times 100,000 \times \text{Weight}_i
  \]
  where `Weight_i` represents the proportion of the population in a given age group.

## **8. How is the data validated to ensure consistency?**
- **Checked for missing values** and handled them based on data type.
- **Verified unique values in categorical columns** (e.g., `age`) to ensure correct classifications.
- **Checked for duplicate rows** and removed them to avoid double counting.
- **Applied outlier detection** by capping extreme values in `estimate`.
